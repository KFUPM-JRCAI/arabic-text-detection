# Build the dataset

```python
dataset = []
dataset.extend(selected_brad_reviews)
dataset.extend(selected_hard_reviews)
len(dataset)
```

```python
dataset[:10]
```

## brief preprocessing

```python
import re
from pyarabic.araby import strip_tatweel

def preprocess(text):
    text = ''.join(char for char in text if char.isprintable())
    
    # Remove tatweel from Arabic text
    text = strip_tatweel(text)
    
    # Handle repeated characters (=, -, *)
    text = re.sub(r'[=\-*]{5,}', '', text)
    
    # Limit repetition of punctuation marks to 3
    text = re.sub(r'[.]{4,}', '...', text)
    text = re.sub(r'[!]{4,}', '!!!', text)
    text = re.sub(r'[?]{4,}', '???', text)
    text = re.sub(r'[_]{3,}', '', text)
    text = re.sub(r'[-]{4,}', '', text)
    text = re.sub(r'[,]{4,}', '', text)
    text = re.sub(r'[،]{4,}', '', text)
    return text
```

```python
preprocessed_dataset = list(map(preprocess,tqdm(dataset)))
preprocessed_dataset[:10]
```

## Brief descriptive statistics

```python
max_words = max(map(lambda sample: len(sample.split()),preprocessed_dataset))
min_words = min(map(lambda sample: len(sample.split()),preprocessed_dataset))
avg_words = sum(map(lambda sample: len(sample.split()),preprocessed_dataset)) / len(preprocessed_dataset)
print(f"max words: {max_words}, min words: {min_words}, avg words: {avg_words}, total samples: {len(preprocessed_dataset)}")
```

## Saving

```python
import json

json.dump(
    preprocessed_dataset,
    open('arabic_datasets/social_media_dataset.json','w'),
    ensure_ascii=False,
    indent=4,
)
```
