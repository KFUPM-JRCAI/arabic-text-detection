```python
# calculate min, max, avg of words in the training set, AI class 1 and human class 0
humans = list(text for (text,label) in data_module.train_dataset if label == 0)
ai = list(text for (text,label) in data_module.train_dataset if label == 1)

humans_words = sum(len(text.split()) for text in humans)
ai_words = sum(len(text.split()) for text in ai)

avg_human_words = humans_words / len(humans)
avg_ai_words = ai_words / len(ai)

min_human_words = min(len(text.split()) for text in humans)
min_ai_words = min(len(text.split()) for text in ai)

max_human_words = max(len(text.split()) for text in humans)
max_ai_words = max(len(text.split()) for text in ai)

print(f'Max: {max(max_human_words,max_ai_words)}',\
f'Min: {max(min_human_words,max_ai_words)},'\
f'avg: {(avg_human_words+avg_ai_words)/2:.2f}\n',\
f'Max words Human: {max_human_words}',\
f'Min words Human: {min_human_words}',\
f'Avg words Human: {avg_human_words:.2f}\n',\
f'Max words AI: {max_ai_words}',\
f'Min words AI: {min_ai_words}',\
f'Avg words AI: {avg_ai_words:.2f}.')
```
